## Processamento de imagem

```python
import cv2
import matplotlib.pyplot as plt
```

```python
import numpy as np
import cv2
frame = cv2.imread('trash_on_water.jpg')
```

```python
frame.shape
```

```python
height, width = frame.shape[:2]
```

### Resize image

Resize image to 1/4 of the original size

```python
resized_frame = cv2.resize(frame, None, fx=0.25, fy=0.25, interpolation=cv2.INTER_AREA)
```

```python
resized_frame.shape
```

```python
plt.imshow(resized_frame)
```

### Turn image to grayscale

```python
gray_frame = cv2.cvtColor(resized_frame, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_frame, cmap='gray')
```

```python
gray_frame.shape
```

### Caculate average of pixel color

```python
h, w = gray_frame.shape[:2]
```

```python
h * w
```

```python
sum_pixels = 0 
for i in gray_frame:
    for j in i:
        sum_pixels += j
```

```python
avg_pixels = sum_pixels / (h*w)
```

```python
avg_pixels
```

### Turn pixels that are off the avg to black

Don't know yet if it is better to remove the pixels that are above or below the avg

```python
# For copy matrix
from copy import copy
```

#### Below the avg

```python
below_avg = copy(gray_frame)
```

```python
for i in range(h):
    for j in range(w):
        if(below_avg[i,j] < avg_pixels):
            below_avg[i,j] = 0
```

```python
below_avg.shape
```

```python
plt.imshow(below_avg, cmap='gray')
```

#### Above the avg

```python
above_avg = copy(gray_frame)
```

```python
for i in range(h):
    for j in range(w):
        if(above_avg[i,j] > avg_pixels):
            above_avg[i,j] = 0
```

```python
above_avg.shape
```

```python
plt.imshow(above_avg, cmap='gray')
```
